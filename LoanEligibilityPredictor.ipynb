```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Bank_Dataset.csv')
df.head()
```

```python
df.info()
```

```python
df['Loan Status'].unique()
```

```python
scale_mapper = {'Loan Refused':0,   
                'Loan Given':1}
```

```python
df['Loan Status Scale'] = df['Loan Status'].replace(scale_mapper)
df.head()
```

```python
df = df.drop(['Loan Status'], axis = 'columns')
df.head()
```

```python
scale_mapper_term = {'Short Term':0,   
                'Long Term':1}
df['Term Scale'] = df['Term'].replace(scale_mapper_term)
df.head()
```

```python
df = df.drop(['Term'], axis = 'columns')
df.head()
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
p = df.hist(figsize = (15,15))
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()


# import seaborn as sns
# sns.heatmap(df.corr(), annot=True)
```
